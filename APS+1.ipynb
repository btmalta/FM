# APS DE F√çSICA DO MOVIMENTO


```python
#EQUA√á√ÉO DO P√äNDULO - IDEAL - 10¬∞

import math
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

#MEDIDO USANDO O TRAKER

periodo = 2.53

t = np.arange(0,periodo,1e-3)
angulo_graus = 10
angulo_rad1 = ((angulo_graus*math.pi)/180)
angulo_graus2 = 45
angulo_rad2 = ((angulo_graus2*math.pi)/180)

g = 9.8
#TAMANNHO DA CORDA
l = 1.7

def Pend1 (Y, tempo):
    o = Y[0]
    w = Y[1]
    dodt = w
    dwdt = (-g*math.sin(o))/l
    return [dodt,dwdt]


Y0 = [angulo_rad1,0]
Y1 = [angulo_rad2,0]

Sol = odeint(Pend1,Y0,t)
Sol2 = odeint(Pend1,Y1,t)

#C√ìDIGO USADO PARA CRIAR O DESLOCAMENTO DO P√äNDULO EM RELA√á√ÉO AO EIXO X E Y

X1 = []
Y1 = []
for i in range(len(Sol[:,0])):
    X1.append(l*(math.sin(Sol[:,0][i])))
    Y1.append(l*(-math.cos(Sol[:,0][i])))
    
#C√ìDIGO USADO PARA CRIAR O DESLOCAMENTO DO P√äNDULO EM RELA√á√ÉO AO EIXO X E Y

X2 = []
Y2 = []
for i in range(len(Sol2[:,0])):
    X2.append(l*(math.sin(Sol2[:,0][i])))
    Y2.append(l*(-math.cos(Sol2[:,0][i])))
    
    
#PLOTANDO O GR√ÅFICO DA MOVIMENTA√á√ÉO DO P√äNDULO EM UM PER√çODO

plt.plot(X1,Y1)
plt.xlabel('Dist√¢ncia em metros')
plt.ylabel('Dist√¢ncia em metros')
plt.title('EQUA√á√ÉO DO P√äNDULO - IDEAL - 10¬∞')
plt.ylim((-1.7,0))
plt.xlim((-1.5,1.5))
plt.grid()
plt.show()
plt.plot(X2,Y2)
plt.xlabel('Dist√¢ncia em metros')
plt.ylabel('Dist√¢ncia em metros')
plt.title('EQUA√á√ÉO DO P√äNDULO - IDEAL - 45¬∞')
plt.ylim((-1.7,0))
plt.xlim((-1.5,1.5))
plt.grid()
plt.show()
```

```python
#GR√ÅFICO DE TEMPO PELO EIXO X - 10¬∞

#TEMPO DIFERENTE DO PER√çODO
T = np.arange(0,60,0.01)

Y0 = [angulo_rad1,0]
Y1 = [angulo_rad2,0]

Sol3 = odeint(Pend1,Y0,T)
Sol4 = odeint(Pend1,Y1,T)

#DESLOCAMENTO PARA O √ÇNGULO DE 10¬∞
X3 = []
Y3 = []
for i in range(len(Sol3[:,0])):
    X3.append(l*(math.sin(Sol3[:,0][i])))
    Y3.append(l*(-math.cos(Sol3[:,0][i])))
    
#DESLOCAMENTO PARA O √ÇNGULO DE 45¬∞
X4 = []
Y4 = []
for i in range(len(Sol4[:,0])):
    X4.append(l*(math.sin(Sol4[:,0][i])))
    Y4.append(l*(-math.cos(Sol4[:,0][i])))
    

plt.plot(T,X3)
plt.xlabel('Tempo')
plt.ylabel('Dist√¢ncia em metros')
plt.title('GR√ÅFICO DE TEMPO PELO EIXO X - 10¬∞')
plt.ylim((-1.5,1.5))
plt.grid()
plt.show()
plt.plot(T,X4)
plt.xlabel('Tempo')
plt.ylabel('Dist√¢ncia em metros')
plt.title('GR√ÅFICO DE TEMPO PELO EIXO X - 45¬∞')
plt.ylim((-1.5,1.5))
plt.grid()

plt.show()

print("COMO NA MODELAGEM N√ÉO FOI UTILIZADO RESIST√äNCIA DO AR, O MOVIMENTO SER√Å INFINITO")
```
